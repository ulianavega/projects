Как мы видим, dividends и stock splits вообще ни с чем не коррелируются, что хорошо для обучения модели.
А вот у Close, Open, High и Low, очевидно, есть очень сильная корреляция. Поэтому земеним их на другие параметры: разность между open и close и разность между high и low.

```python
#создаем столбик High - Low
amzn['High - Low'] = amzn['High'] - amzn['Low']
google['High - Low'] = google['High'] - google['Low']
apple['High - Low'] = apple['High'] - apple['Low']
msft['High - Low'] = msft['High'] - msft['Low']

#теперь создаем Open - Close
amzn['Open - Close'] = amzn['Open'] - amzn['Close']
google['Open - Close'] = google['Open'] - google['Close']
apple['Open - Close'] = apple['Open'] - apple['Close']
msft['Open - Close'] = msft['Open'] - msft['Close']
```

Попробуем предсказать, нужно ли покупать акции Амазона, то есть решим проблему классификации: 1 - нужно покупать, 0 - не нужно \\
Используем XGBoost


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn import metrics
```

```python
#покупаем, если 1. Вообще 1 - это индикатор того, что завтра цена вырастет

amzn['target'] = np.where(amzn['Close'].shift(-1) > amzn['Close'], 1, 0)
```

```python
#подготавливаем данные. Для этого 10% данных оставляем для теста

features = amzn[['High - Low', 'Open - Close', 'Volume', 'Stock Splits']]
target = amzn['target']

scaler = StandardScaler()
features = scaler.fit_transform(features)

X_train, X_valid, Y_train, Y_valid = train_test_split(
    features, target, test_size=0.1, random_state=2022)
print(X_train.shape, X_valid.shape)
print(Y_train.shape, Y_valid.shape)
Y_train.head()
```

```python
#обучим и оценим точность

model = XGBClassifier()
model.fit(X_train, Y_train)

print('Training Accuracy : ', metrics.roc_auc_score(Y_train, model.predict_proba(X_train)[:,1]))
print('Validation Accuracy : ', metrics.roc_auc_score(Y_valid, model.predict_proba(X_valid)[:,1]))
```

Как мы видим, точность модели, как и вероятность угадать - 50%. Поэтому ее надо улучшить. Для начала попробуем другие модели

```python
model = LogisticRegression()
model.fit(X_train, Y_train)

print('Training Accuracy : ', metrics.roc_auc_score(Y_train, model.predict_proba(X_train)[:,1]))
print('Validation Accuracy : ', metrics.roc_auc_score(Y_valid, model.predict_proba(X_valid)[:,1]))
```

```python
model = SVC(kernel='poly', probability=True)
model.fit(X_train, Y_train)

print('Training Accuracy : ', metrics.roc_auc_score(Y_train, model.predict_proba(X_train)[:,1]))
print('Validation Accuracy : ', metrics.roc_auc_score(Y_valid, model.predict_proba(X_valid)[:,1]))
```

Как видим, с ними дела обстоят еще хуже. Попробуем добавить больше парметров,
которые могут повлиять на цену. Попробуем учитывать цену на акции аналогичных компаний (те, что мы использовали, кроме Амазона)

Это толкьо ухудшило ситуацию. Попробуем другую модель

```python
from fastai.tabular.all import *

splits = RandomSplitter(valid_pct=0.2)(range_of(amzn))
```

```python
to = TabularPandas(amzn,
                   cont_names = ['High - Low', 'Open - Close', 'Volume', 'Stock Splits'],
                   y_names='target', procs = [Categorify, FillMissing, Normalize], splits = splits)

dls = to.dataloaders(bs=64)
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=accuracy)
learn.fit_one_cycle(1)
```

```python
learn.show_results()
```
