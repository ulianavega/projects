```python
sns.histplot(data=data, x='YEAR', bins = 7, color = 'lightskyblue')
plt.title('Количество полетов каждый год')
plt.xlabel('Год')
plt.ylabel('Частота')
```

```python
sns.histplot(data['MONTH_MON'][data['YEAR'] != 2022], bins = 77, color = 'lightskyblue')
plt.title('Количество полетов каждый месяц')
plt.xlabel('Месяц')
plt.ylabel('Частота')
```

```python
sns.histplot(data['MONTH_MON'][data['YEAR'] != 2022], bins = 77, color = 'lightskyblue', kde=True)
plt.title('Количество полетов каждый месяц')
plt.xlabel('Месяц')
plt.ylabel('Частота')
```

```python
sns.histplot(data['FLT_DATE'][(data['YEAR'] == 2018) & (data['MONTH_MON'] == 'JAN')], bins = 77, color = 'lightskyblue', kde=True)
plt.title('Количество полетов в январе 2018')
plt.xlabel('Месяц')
plt.xticks(rotation = 90)
plt.ylabel('Частота')
```

```python
sns.histplot(data['MONTH_MON'][(data['STATE_NAME'] == 'Ukraine') & (data['YEAR'] == 2022)], bins = 7, color = 'orange')
plt.title('Статистика полётов в Украине')
plt.xlabel('Год')
plt.ylabel('Частота')
```
